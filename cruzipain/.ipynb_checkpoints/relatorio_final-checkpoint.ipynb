```python
import pandas as pd
import numpy as np
import os
```

```python
resultados = {}
for item in os.popen("ls fukui/*.csv").read().split('\n')[:-1]:
    if "mulli" in item:
        if "alpha_b_un_" in item:
            resultados[item.removeprefix("fukui/fukui_mulliken_").removesuffix(".csv")[:-3].replace("alpha_b_un_", "alpha_beta_unsaturated_")] = pd.read_csv(item)
        else:
            resultados[item.removeprefix("fukui/fukui_mulliken_").removesuffix(".csv")[:-3]] = pd.read_csv(item)
resultados.keys()
```

```python
len(resultados)
```

```python
df = pd.read_csv('fukui/flig.smi',sep=' ')
df["mulliken_fk+"] = np.nan
df["mulliken_fk-"] = np.nan
df["mulliken_fk0"] = np.nan
df
```

```python
for key in list(resultados.keys()):
    index_warhead, index_df = df.query("NAME == @key").WARHEAD, df.query("NAME == @key").index[0]
    df.iloc[index_df, 3] = resultados[key].loc[index_warhead]['fk+']
    df.iloc[index_df, 4] = resultados[key].loc[index_warhead]['fk-']
    df.iloc[index_df, 5] = resultados[key].loc[index_warhead]['fk0']
```
